## Introduction

The background for the problem originates from the multitude of online forums, wherein people participate actively and make comments. Conversational toxicity on online platforms is a huge issue that can lead people to stop expressing themselves and to stop seeking others' opinions out of fear of abuse or harassment. For an online forum, identifying toxicity of comments becomes very important to put a check on users from posting potentially hurtful comments, to craft more civil arguments when engaging in social discourse with others, and to gauge the toxicity of users' comments. The goal is to build a tool that classifies comments into various categories and helps improve the online conversation.
<br>
<br>
![title](images/thumbnail_tcc_2.jpg)

<br>

## Objective

Given a group of sentences or paragraphs, used as a comment on an online platform, classify it to belong to one or more of the following categories, such as __toxic, severe-toxic, obscene, threat, insult, or identity-hate__. A model has to be created that predicts the probability of each type of toxicity for a given comment.

## Dataset

A few details regarding the dataset are as follows:
1. It comes from __Jigsaw's Toxic Comment Classification__ challenge on __Kaggle__.
2. It consists of __159,571__ labeled training examples of __Wikipedia__ comments that have been labeled by human raters for toxic behavior and __153,164__ unlabeled test comments.
3. The types of toxicity are:
    1. Toxic
    2. Severe-toxic
    3. Obscene
    4. Threat
    5. Insult
    6. Identity-hate
4. The data comes in the schema of <id, comment_text, toxic, severe_toxic, obscene, threat, insult, identity_hate>, where the labels for _toxic, severe_toxic, obscene, threat, insult, and identity_hate_ are all boolean.

* get the base form of words using lemmatization

```python
# this function contain the code for text cleaning.
tokenizer=TweetTokenizer()
lem = WordNetLemmatizer()
def do_cleaning(x, contraction_map):
    x = str(x)
    x= x.lower()
    x= re.sub(r'https?:\/\/[^ ]*', '', x) #to remove any urls
    x= re.sub(r'\d+.\d+.\d+.\d+', '', x)  #to remove any ip
    x= re.sub(r'\[\[.*\]', '', x)      #to remove any username enclosed in []
    
    for punct in "/-'":
        x = x.replace(punct, ' ')
    for punct in '&':
        x = x.replace(punct, f' {punct} ')
    for punct in '?!.,"#$%\'()*+-/:;<=>@[\\]^_`{|}~' + '“”’\n':
        x = x.replace(punct, '')
    x= re.sub(r' +', r' ', x)
    x= x.strip(' ')

    words=tokenizer.tokenize(x)
    
    # (')aphostophe  replacement (ie)   you're --> you are  
    # ( basic dictionary lookup : master dictionary present in a hidden block of code)
    words=[contraction_map[word] if word in contraction_map else word for word in words]
    words=[lem.lemmatize(word, "v") for word in words]
    words = [w for w in words if not w in stop_words]
    
    clean_sent=" ".join(words)
    return clean_sent

train['comment_text_clean']= train['comment_text'].apply(lambda x: do_cleaning (x, contraction_mapping))
test['comment_text_clean']= test['comment_text'].apply(lambda x: do_cleaning (x, contraction_mapping))
```

```python
#This function calculates the coverage/overlap of corpus vocabulary with different embeddings.
def calc_coverage(vocab, embedding, embed_type):
    vocab_words= set(vocab.keys())
    embed_words= set(embedding.keys())
    common_words= vocab_words.intersection(embed_words)
    total_words_match= 0
    for word in common_words:
        total_words_match+=vocab[word]
    oov_words= vocab_words - common_words
    oov_words_cnt= []
    for word in oov_words:
        oov_words_cnt.append((word,vocab[word]))
    oov_words_cnt.sort(key= itemgetter(1), reverse= True)
    total_corpus_len= np.sum(list(vocab.values()))
    pcntg_common_words= (len(common_words)/len(vocab_words))*100
    total_pcntg_words_match= (total_words_match/total_corpus_len)*100
    display(Markdown("_Total number of words in {} Embedding:_ __{}__".format(embed_type, len(embed_words))))
    display(Markdown("_Total number of common words between vocabulary and {0} Embedding:_ __{1}__".format(embed_type, len(common_words))))
    display(Markdown("_Percentage of common words between vocabulary and {0} Embedding:_ __{1:.2f}__".format(embed_type, pcntg_common_words)))
    display(Markdown("_Percentage of common words between our corpus and {0} Embedding:_ __{1:.2f}__".format(embed_type, total_pcntg_words_match)))
#     print("Total number of out of vocab words for {} Embedding: {}".format(embed_type, len(oov_words_cnt.keys())))                                                                                     
#     return oov_words_cnt
    return pcntg_common_words, total_pcntg_words_match
```

We have considered three different embeddings which are as follows:
* GloVe

```python
## Indirect features

#Sentense count in each comment:
    #  '\n' can be used to count the number of sentences in each comment
df['count_sent']=df["comment_text"].apply(lambda x: len(re.findall("\n",str(x)))+1)
#Word count in each comment:
df['count_word']=df["comment_text"].apply(lambda x: len(str(x).split()))
#Unique word count
df['count_unique_word']=df["comment_text"].apply(lambda x: len(set(str(x).split())))
#Letter count
df['count_letters']=df["comment_text"].apply(lambda x: len(str(x)))
#punctuation count
df["count_punctuations"] =df["comment_text"].apply(lambda x: len([c for c in str(x) if c in string.punctuation]))
#upper case words count
df["count_words_upper"] = df["comment_text"].apply(lambda x: len([w for w in str(x).split() if w.isupper()]))
#title case words count
df["count_words_title"] = df["comment_text"].apply(lambda x: len([w for w in str(x).split() if w.istitle()]))
#Number of stopwords
df["count_stopwords"] = df["comment_text"].apply(lambda x: len([w for w in str(x).lower().split() if w in stop_words]))
#Average length of the words
df["mean_word_len"] = df["comment_text"].apply(lambda x: np.mean([len(w) for w in str(x).split()]))
#derived features
#Word count percent in each comment:
df['word_unique_percent']=df['count_unique_word']*100/df['count_word']
#derived features
#Punct percent in each comment:
df['punct_percent']=df['count_punctuations']*100/df['count_word']
```

```python
def top_tfidf_feats(row, features, top_n=25):
    ''' Get top n tfidf values in row and return them with their corresponding feature names.'''
    topn_ids = np.argsort(row)[::-1][:top_n]
    top_feats = [(features[i], row[i]) for i in topn_ids]
    df = pd.DataFrame(top_feats)
    df.columns = ['feature', 'tfidf']
    return df

def top_feats_in_doc(Xtr, features, row_id, top_n=25):
    ''' Top tfidf features in specific document (matrix row) '''
    row = np.squeeze(Xtr[row_id].toarray())
    return top_tfidf_feats(row, features, top_n)

def top_mean_feats(Xtr, features, grp_ids, min_tfidf=0.1, top_n=25):
    ''' Return the top n features that on average are most important amongst documents in rows
        indentified by indices in grp_ids. '''
    
    D = Xtr[grp_ids].toarray()

    D[D < min_tfidf] = 0
    tfidf_means = np.mean(D, axis=0)
    return top_tfidf_feats(tfidf_means, features, top_n)

# modified for multilabel milticlass
def top_feats_by_class(Xtr, features, min_tfidf=0.1, top_n=20):
    ''' Return a list of dfs, where each df holds top_n features and their mean tfidf value
        calculated across documents with the same class label. '''
    dfs = []
    cols=train_tags.columns
    for col in cols:
        ids = train_tags.index[train_tags[col]==1]
        feats_df = top_mean_feats(Xtr, features, ids, min_tfidf=min_tfidf, top_n=top_n)
        feats_df.label = label
        dfs.append(feats_df)
    return dfs
```

### Modeling

A few modeling details are as follows:
* Modeling has been done with __Bidirectional GRU__ units, __LSTM__ has also been tried. Although both LSTM and GRU perform rather the same in terms of accuracy, GRU trains faster for obvious reasons.
* The bidirectional version of GRU is found to give better results.
* Different pre-trained embeddings __(Wiki-news-300, Google-news-300, and Glove-300)__ and their combinations have been tried but the concatenation of Wiki-news and Google-news is found to produce better results.
* Pre-trained embeddings are not fine-tuned as part of the training process.
* Apart from pre-trained embeddings; meta-features, as discussed in the feature engineering section, have also been used which slightly improves the accuracy.
* Vocabulary size and the maximum length of a comment are restricted to __30000__ and __200__ respectively.
* __Snapshot Ensembles__ has also been used to train the model better and achieve better predictions. 
7. Below are the hyper-parameters which have been obtained after hyper-parameter tuning:

    * __Number of bi-directional GRU layers:__ 2
    2. __Number of GRU units in a layer:__ 128
    3. __Number of hidden dense layers:__ 2
    4. __Number of hidden dense units:__ 512
    5. __Batch size:__ 512
    6. __Number of epochs:__ 4 with __early stopping__
    7. __Spatial Dropout after embedding layer:__ 0.2
    8. __Dropout after hidden dense layers:__ 0.1
    9. __Optimizer:__ Adam
    
8. Achieved __mean column-wise AUC__ score of __0.986__.  

```python
SELECTED_COLS=['count_sent', 'count_word', 'count_unique_word',
       'count_letters', 'count_punctuations', 'count_words_upper',
       'count_words_title', 'count_stopwords', 'mean_word_len',
       'word_unique_percent', 'punct_percent']
# target_x=train_feats[SELECTED_COLS]
TARGET_COLS=['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
# target_y=train_feats[TARGET_COLS]
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()

# classes= ['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']
train_df, val_df = train_test_split(train_feats, test_size=0.1)
train_text_X= train_df['comment_text_clean'].values
val_text_X= val_df['comment_text_clean'].values
test_text_X= test_feats['comment_text_clean'].values
train_meta_X= train_df[SELECTED_COLS].values
val_meta_X= val_df[SELECTED_COLS].values
test_meta_X= test_feats[SELECTED_COLS].values
scaler.fit(train_meta_X)
train_meta_X= scaler.transform(train_meta_X)
val_meta_X= scaler.transform(val_meta_X)
test_meta_X= scaler.transform(test_meta_X)
train_y= train_df[TARGET_COLS].values
val_y= val_df[TARGET_COLS].values

# print( train.shape, test.shape,train.columns, test.columns)
# # print(train_df['target'].value_counts()[0]/train_df['target'].value_counts()[1])
# # print(val_df['target'].value_counts()[0]/val_df['target'].value_counts()[1])
```

```python
## Tokenize the sentences
tokenizer = Tokenizer(num_words=vocab_size)
tokenizer.fit_on_texts(list(train_text_X))
train_text_X = tokenizer.texts_to_sequences(train_text_X)
val_text_X = tokenizer.texts_to_sequences(val_text_X)
test_text_X = tokenizer.texts_to_sequences(test_text_X)

## Pad the sentences 
train_text_X = pad_sequences(train_text_X, maxlen=maxlen)
val_text_X = pad_sequences(val_text_X, maxlen=maxlen)
test_text_X = pad_sequences(test_text_X, maxlen=maxlen)
```

```python
def extend_embed(embed_dict, vocab_size):
    all_embs = np.stack(embed_dict.values())
    emb_mean,emb_std = all_embs.mean(), all_embs.std()
    embed_size = all_embs.shape[1]
    word_index = tokenizer.word_index
    nb_words = min(vocab_size, len(word_index))
    embedding_matrix = np.random.normal(emb_mean, emb_std, (nb_words, embed_size))
    for word, i in word_index.items():
        if i >= vocab_size: continue
        embedding_vector = embed_dict.get(word)
        if embedding_vector is not None: embedding_matrix[i] = embedding_vector
    return embedding_matrix

glove_embed_mat= extend_embed(glove_dict, vocab_size)
wiki_embed_mat= extend_embed(wiki_dict, vocab_size)
google_embed_mat= extend_embed(google_dict, vocab_size)

embedding_matrix= np.concatenate((glove_embed_mat, wiki_embed_mat), axis=1)
# embedding_matrix= np.mean((glove_embed_mat, wiki_embed_mat, google_embed_mat), axis=0)
# embedding_matrix= glove_embed_mat
```

```python
def build_model(embedding_matrix, use_meta_feat):
    embed_size= embedding_matrix.shape[1]
    n_meta_feat= train_meta_X.shape[1]
    inp = Input(shape=(maxlen,))
    x = Embedding(vocab_size, embed_size, weights=[embedding_matrix], trainable= False)(inp)
    #spatial dropout is used to drop several random dimensions from all the word embeddings within a sentence.
    x = SpatialDropout1D(0.2)(x)
    x = Bidirectional(GRU(LSTM_UNITS, return_sequences=True))(x)
    x = Bidirectional(GRU(LSTM_UNITS, return_sequences=True))(x)
    avg_pool = GlobalAveragePooling1D()(x)
    max_pool = GlobalMaxPooling1D()(x)
    meta_inp= Input(shape= (n_meta_feat,))
    if(use_meta_feat== True):
        x = concatenate([avg_pool, max_pool, meta_inp])
    else:
        x = concatenate([avg_pool, max_pool])
    x = Dense(DENSE_HIDDEN_UNITS, activation="relu")(x)
    x = Dense(DENSE_HIDDEN_UNITS, activation="relu")(x)
    x = Dropout(0.1)(x)
    x = Dense(n_class, activation="sigmoid")(x) #we are using sigmoid, instead of softmax, since this is multilabel classification.
    if(use_meta_feat== True):
        model = Model(inputs=[inp, meta_inp], outputs=x)
    else:
        model = Model(inputs=inp, outputs=x)
    #in multilabel classification, we use sigmoid with binary_crossentropy.
    #in multiclass classification, we use softmax with categorical_crossentropy.
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
checkpoint_predictions = []
weights = []
model = build_model(embedding_matrix, use_meta_feat)
history_all= []
for model_idx in range(NUM_MODELS):
    for global_epoch in range(EPOCHS):
        if(use_meta_feat==False):
            history= model.fit(train_text_X, train_y, batch_size= BATCH_SIZE, epochs=1, verbose=0, validation_data=(val_text_X, val_y),callbacks=[LearningRateScheduler(lambda _: init_lr * (lr_scaling_factor ** global_epoch))])
            checkpoint_predictions.append(model.predict(val_text_X, batch_size= BATCH_SIZE))
        else:
            history= model.fit([train_text_X,train_meta_X], train_y, batch_size=BATCH_SIZE, epochs=1, verbose=0, validation_data=([val_text_X, val_meta_X], val_y),callbacks=[LearningRateScheduler(lambda _: init_lr * (lr_scaling_factor ** global_epoch))])
            checkpoint_predictions.append(model.predict([val_text_X, val_meta_X], batch_size=BATCH_SIZE))
        weights.append(2 ** global_epoch)
        history_all.append(history)
predictions = np.average(checkpoint_predictions, weights=weights, axis=0)
```
